```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
NTBK_DIR = os.path.dirname(cwd)

DATA_DIR = os.path.join(NTBK_DIR, 'box_office_data')

CACHE_DIR = os.path.join(NTBK_DIR, 'cache')
os.makedirs(CACHE_DIR, exist_ok=True)
```

```python
for f in os.listdir(DATA_DIR):
    print(f)
```

```python
movie_dataframes = []

#isolate csv files
csv_files = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]

#iterate over the CSV files
for filename in os.listdir(DATA_DIR):
    # strip .csv for year 
    year = filename.replace(".csv", "")
    # get file path
    csv_path = os.path.join(DATA_DIR, filename)
    
    # read csv file
    this_df = pd.read_csv(csv_path)
    
    # update row names
    this_df['filename'] = filename
    this_df['year'] = year

    # append to list
    movie_dataframes.append(this_df)
```

```python
entire_df = pd.concat(movie_dataframes)
entire_df.head()
```

```python
entire_df.tail()
```

```python
# extract entire dataframe and create a new CSV
dataset = os.path.join(CACHE_DIR, "box_office_dataset.csv")
entire_df.to_csv(dataset, index=False)
```
